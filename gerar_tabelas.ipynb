```python
import pandas as pd
import numpy as np
import datetime
from datetime import timedelta, datetime
from faker import Faker
import random
import string
```

#Funções

```python
def quantidade_linhas_status(df, coluna):
    lista_status = sorted(df[coluna].unique())

    for status in lista_status:
      print(f"{status}: {len(df[df[coluna] == status])}")
```

```python
def generate_commit_date(dt_deploy):
    max_commit_days_difference = 7
    max_commit_date = dt_deploy - timedelta(days=max_commit_days_difference)
    return fake.date_time_between(start_date=max_commit_date, end_date=dt_deploy)
```

```python
def generate_opened_at_date(closed_at):
    max_opened_at_days_difference = 3
    max_opened_at_date = closed_at - timedelta(days=max_opened_at_days_difference)
    return fake.date_time_between(start_date=max_opened_at_date, end_date=closed_at)
```

```python
def generate_caller_id():
    return ''.join(random.choices(string.ascii_letters + string.digits, k=8))
```

#**Esteira de Produção**

```python
# Definindo a data inicial e final do intervalo
data_inicial = pd.to_datetime('2022-01-01')
data_final = pd.to_datetime('2022-12-31')

# Criando um gerador de dados aleatórios para os campos
fake = Faker()

# Lista para armazenar os dados gerados
data_list = []

# Definindo as opções de status e suas respectivas probabilidades
status_options = ['Sucesso', 'Falha']
status_probabilities = [0.80, 0.20]  # Probabilidades: Sucesso (80%), Falha (20%)

# Gerando 30 linhas de dados
for numero in range(1, 121):  # Números de 1 a 120
    # Gerando os dados aleatórios para cada campo
    id_deploy = fake.uuid4()
    time = random.choice(['Team Alpha', 'Team Beta', 'Team Gamma'])
    status = random.choices(status_options, weights=status_probabilities)[0]
    dt_deploy = fake.date_time_between(start_date=data_inicial, end_date=data_final)
    dt_commit = generate_commit_date(dt_deploy)

    #Swfinindo a atuação a partir do time
    if time == 'Team Alpha':
        atuacao = 'DevEng'
    elif time == 'Team Beta':
        atuacao = 'DevOps'
    else:
        atuacao = 'UX'

    # Adicionando a linha de dados à lista
    data_list.append([id_deploy, time, atuacao, numero, status, dt_commit, dt_deploy])

# Convertendo a lista para um DataFrame
esteira_prod = pd.DataFrame(data_list, columns=['idDeploy', 'time', 'atuacao', 'funcionalidade', 'status', 'dt_commit', 'dt_deploy'])

esteira_prod
```

```python
quantidade_linhas_status(esteira_prod, 'status')
```

##Salvando como CSV

```python
esteira_prod.to_csv('esteira_prod.csv', index = False)
```

#**Chamados**

```python
# Definindo a data inicial e final do intervalo
data_inicial = pd.to_datetime('2022-01-01')
data_final = pd.to_datetime('2022-12-31')

# Criando um gerador de dados aleatórios para os campos
fake = Faker()

# Criando um gerador de números aleatórios
random.seed(42)

# Lista para armazenar os dados gerados
data_list = []

# Definindo as categorias e códigos de u_symptom
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
u_symptoms = ['SYM001', 'SYM002', 'SYM003', 'SYM004', 'SYM005']

# Gerando a tabela de chamados
for _ in range(1, 587):  # Gerar 100 chamados
    # Gerando os dados aleatórios para cada campo
    call_id = random.randint(100000, 999999)
    caller_id = random.randint(1000, 10000)
    category = random.choice(categories)
    u_symptom = random.choice(u_symptoms)
    assignment_group = f"Group {random.randint(1, 5)}"
    incident_state = random.choices(['closed'], weights=[0.1])[0]
    closed_at = fake.date_time_between(start_date=data_inicial, end_date=data_final)
    opened_at = generate_opened_at_date(closed_at)

    # Adicionando a linha de dados à lista
    data_list.append([call_id, caller_id, category, u_symptom, assignment_group, incident_state, opened_at, closed_at])

# Convertendo a lista para um DataFrame
chamados = pd.DataFrame(data_list, columns=['id', 'caller_id', 'category', 'u_symptom', 'assignment_group', 'incident_state', 'opened_at', 'closed_at'])

# Exibindo a tabela de chamados
chamados
```

```python
quantidade_linhas_status(chamados, 'incident_state')
```

##Salvando como CSV

```python
chamados.to_csv('chamados.csv', index = False)
```

#**Log**

```python
# Definindo a data inicial e final do intervalo
data_inicial = pd.to_datetime('2022-01-01')
data_final = pd.to_datetime('2022-12-31')

# Criando um gerador de dados aleatórios para os campos
fake = Faker()

# Lista de erros de exemplo para o serviço PAAS (erros HTTP)
service_errors = [
    "400 Bad Request",
    "401 Unauthorized",
    "403 Forbidden",
    "404 Not Found",
    "500 Internal Server Error",
    "502 Bad Gateway",
    "503 Service Unavailable",
]

# Lista para armazenar os dados gerados
data_list = []

# Gerando a tabela de logs de serviço
for _ in range(100):  # Gerar 100 logs
    # Gerando os dados aleatórios para cada campo
    id_log = fake.uuid4()

    # Gerando o timestamp_up no ano de 2022
    timestamp_up = fake.date_time_between(start_date=data_inicial, end_date=data_final)

    # Gerando o timestamp_down até 12 horas antes do timestamp_up
    timestamp_down = timestamp_up - timedelta(hours=random.randint(1, 12))

    # Gerando um erro aleatório de servidor (erro HTTP)
    error = random.choice(service_errors)

    # Adicionando a linha de dados à lista
    data_list.append([id_log, timestamp_down, timestamp_up, error])

# Convertendo a lista para um DataFrame
logs = pd.DataFrame(data_list, columns=['idLog', 'timestamp_down', 'timestamp_up', 'error'])

# Exibindo a tabela de logs de serviço
logs
```

```python
logs.sort_values(by='timestamp_down', ascending=False)
```

##Salvando como CSV

```python
logs.to_csv('logs.csv', index = False)
```
